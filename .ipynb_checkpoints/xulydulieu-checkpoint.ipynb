# Trực quan hóa dữ liệu

```python
df = pd.read_csv('heart_Ok.csv')
df.info()
```

```python
# Tạo biểu đồ phân phối với đường giá trị trung bình cho các cột 'age', 'trestbps', 'chol'
columns_to_plot = ['age', 'chol']
averages = df[columns_to_plot].mean()

# Thiết lập biểu đồ
plt.figure(figsize=(15, 15))

for i, col in enumerate(columns_to_plot, 1):
    plt.subplot(2,1, i)
    sns.histplot(df[col], kde=True, color='blue', bins=20)
    plt.axvline(averages[col], color='red', linestyle='--', label=f'Mean: {averages[col]:.2f}')
    plt.title(f'Biểu đồ phân phối của {col}')
    plt.xlabel(col)
    plt.ylabel('Tần suất')
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Ma trận tương quan
plt.figure(figsize=(10, 8))
corr = df.corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", cbar=True, square=True)
plt.title("Ma trận tương quan")
plt.show()
```

```python
# Biểu đồ 1: Phân bố giá trị 'target' 
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
# Biểu đồ cột
sns.countplot(data=df, x='target', palette='viridis', ax=axes[0])
axes[0].set_title("Phân bố kết quả bệnh tim", fontsize=14)
axes[0].set_xlabel("Kết quả (1: Bệnh, 0: Không bệnh)", fontsize=12)
axes[0].set_ylabel("Số lượng", fontsize=12)

# Biểu đồ tròn
data['target'].value_counts().plot(kind='pie', autopct='%1.1f%%', labels=['Bệnh', 'Không bệnh'], colors=['#1f77b4', '#ff7f0e'], startangle=90, explode=(0.1, 0), ax=axes[1])
axes[1].set_title("Tỷ lệ bệnh tim theo kết quả", fontsize=14)
axes[1].set_ylabel("")

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Tạo một hình với kích thước cụ thể
plt.figure(figsize=(8, 6))

# Vẽ biểu đồ cột với seaborn, sử dụng cột 'target' trên trục x và cột 'cp' trên trục y
sns.barplot(data=df, x='target', y='cp', palette='viridis',ci=None)

# Thêm tiêu đề cho biểu đồ
plt.title("Loại đau ngực (CP) theo kết quả bệnh tim", fontsize=14)

# Thêm nhãn cho trục x và trục y
plt.xlabel("Kết quả (1: Bệnh, 0: Không bệnh)", fontsize=12)
plt.ylabel("Loại đau ngực (CP)", fontsize=12)

# Hiển thị biểu đồ
plt.show()
```

```python
# Biểu đồ bổ sung 1: Quan hệ giữa 'cp' và 'target'
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='target', y='exang', palette='viridis',ci=None)
plt.title("Loại đau ngực (CP) theo kết quả bệnh tim", fontsize=14)
plt.xlabel("Kết quả (1: Bệnh, 0: Không bệnh)", fontsize=12)
plt.ylabel("Loại đau ngực (CP)", fontsize=12)
plt.show()
```

```python
# Biểu đồ bổ sung 1: Quan hệ giữa 'cp' và 'target'
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='target', y='ca', palette='viridis',ci=None)
plt.title("Loại đau ngực (CP) theo kết quả bệnh tim", fontsize=14)
plt.xlabel("Kết quả (1: Bệnh, 0: Không bệnh)", fontsize=12)
plt.ylabel("Loại đau ngực (CP)", fontsize=12)
plt.show()
```

```python
# Biểu đồ bổ sung 2: Quan hệ giữa 'thalach' và 'target'
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='target', y='thalach', palette='cool')
plt.title("Nhịp tim tối đa (Thalach) theo kết quả bệnh tim", fontsize=14)
plt.xlabel("Kết quả (1: Bệnh, 0: Không bệnh)", fontsize=12)
plt.ylabel("Nhịp tim tối đa (Thalach)", fontsize=12)
plt.show()
```

```python
# Tính Q1 (25%) và Q3 (75%)
Q1 = df['thalach'].quantile(0.25)
Q3 = df['thalach'].quantile(0.75)
IQR = Q3 - Q1

# Xác định ngưỡng dưới và ngưỡng trên
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5* IQR

# Loại bỏ các giá trị ngoại lai
data_cleaned = df[(df['thalach'] >= lower_bound) & (df['thalach'] <= upper_bound)]
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=data_cleaned, x='target', y='thalach', palette='cool')
plt.title("Nhịp tim tối đa (Thalach) theo kết quả bệnh tim", fontsize=14)
plt.xlabel("Kết quả (1: Bệnh, 0: Không bệnh)", fontsize=12)
plt.ylabel("Nhịp tim tối đa (Thalach)", fontsize=12)
plt.show()
```
